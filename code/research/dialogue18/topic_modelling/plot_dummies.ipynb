### Amount of papers for different years of Dialogue

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height, '%d' % int(height), ha='center', va='bottom', fontsize=20)
```

### Amount of papers for different years of Dialogue

```python
x = [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]

plt.figure(figsize=(18, 8))
plt.xticks(x)
plt.tick_params(labelsize=20)

line1 = plt.plot(x, [0, 0, 0, 0, 0, 10, 11, 14, 8, 12, 11], 
                 label='AIST', linewidth=4, marker='^', markersize=15)
line2 = plt.plot(x, [0, 0, 0, 0, 0, 0, 0, 0, 34, 24, 23], 
                 label='AINL', linewidth=4, marker='o', markersize=15)
line3 = plt.plot(x, [100, 97, 96, 98, 73, 88, 90, 65, 87, 100, 99],
                 label='Dialogue', linewidth=4, marker='<', markersize=15)


plt.grid(True, axis='y', linewidth=1, color='grey')

leg = plt.legend(loc='best', prop={'size':30}, frameon=True)
leg.get_frame().set_linewidth(3)
leg.get_frame().set_edgecolor('black')

plt.show()
# plt.savefig('distribution.png', bbox_inches='tight')
```

```python
sns.set_style('white')

N = 6
men_means = [88, 90, 65, 87, 100, 99]

ind = np.arange(N)  
width = 0.2       

fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(15)
rects1 = ax.bar(ind, men_means, width, color='r')

women_means = [10, 11, 14, 8, 12, 11]
rects2 = ax.bar(ind + width, women_means, width, color='y')

kea_means = [0, 0, 0, 30, 24, 23]
rects3 = ax.bar(ind + width*2, kea_means, width, color='green')

ax.set_ylabel('Amount', fontsize=20)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('2012', '2013', '2014', '2015', '2016', '2017'), fontsize=14)
ax.legend((rects1[0], rects2[0], rects3[1]), ('Dialogue', 'AIST', 'AINL'))

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.tick_params(which='both', labelsize=20)
plt.show()
```
