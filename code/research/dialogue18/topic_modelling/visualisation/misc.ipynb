```python
import pyLDAvis.gensim
import gensim
from gensim import corpora
from basic_functions import _get_data_from_cmd
from basic_functions import _parse_files
import pickle
from gensim import models
```

```python
 with open('data/pos_lemmas_with_normalized_words.pickle', 'rb') as handle:
        all_texts_dict = pickle.load(handle)
```

```python
top_topics = {}
dictionary = corpora.Dictionary(list(all_texts_dict.values()))
dictionary.filter_n_most_frequent(20)
```

```python
ldamodel = gensim.models.ldamodel.LdaModel.load("created_models/lda_model_freq0_topics_10_tfidf_True_pos_False_norm_True/lda_model_freq0_topics_10_tfidf_True_pos_False_norm_True.mdl")
```

```python
corpus = [dictionary.doc2bow(doc) for doc in list(all_texts_dict.values())]
tfidf = models.TfidfModel(corpus)
corpus = tfidf[corpus]
```

```python
ldamodel.log_perplexity(corpus)
```

```python
visdata = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
```

```python
pyLDAvis.display(visdata)
```

```python
def byName_key(topic):
    return topic[1]
```

```python
vals = list(all_texts_dict.values())
names = list(all_texts_dict.keys())

all_topics = dict(ldamodel.print_topics(num_topics=50, num_words=5))
for a in all_topics:
    print(a,": ",  all_topics[a])
print()

for i in range(5):
    print(list(reversed(sorted(ldamodel.get_document_topics(dictionary.doc2bow(vals[i])), key=byName_key)))[:3], names[i])
```
