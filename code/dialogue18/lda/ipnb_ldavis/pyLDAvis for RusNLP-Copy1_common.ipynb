```python
frq = 0
topics = 10
raw = 0

if raw == 1:
    data = all_texts_dict_raw
elif raw == 0:
    data = all_texts_dict_pos

data = list(data.values())
```

```python
ldamodel = gensim.models.ldamodel.LdaModel.load("new_output/lda_model_normalized_words_freq__common_"+str(frq)+"_topics_"+str(topics)+"_pos_"+str(raw)+"/lda_model_normalized_words_freq__common_"+str(frq)+"_topics_"+str(topics)+"_pos_"+str(raw)+"_common_.mdl")
print("/lda_model_normalized_words_freq__common_"+str(frq)+"_topics_"+str(topics)+"_pos_"+str(raw)+"_common_.mdl")
```

```python
top_topics = {}
dictionary = corpora.Dictionary(data)
dictionary.filter_n_most_frequent(frq)
corpus = [dictionary.doc2bow(doc) for doc in data]
```

```python
ldamodel.log_perplexity(corpus)
```

```python
visdata = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
```

```python
vals = list(all_texts_dict_pos.values())
names = list(all_texts_dict_pos.keys())

all_topics = dict(ldamodel.print_topics(num_topics=topics, num_words=5))
for a in all_topics:
    print(a,": ",  all_topics[a])
print()

#names[i] -- сюда можно вставить айдишник текста и проверить на правильность
for i in range(5):
    print(list(reversed(sorted(ldamodel.get_document_topics(dictionary.doc2bow(vals[i])), key=byName_key)))[:3], names[i])
```

```python
frq = [0,20]
topics = [10, 50]
raw = [0,1]

for fr in frq:
    for topic in topics:
        for pos_raw in raw:
            if pos_raw == 1:
                data = list(all_texts_dict_raw.values())
                old_data = all_texts_dict_raw
            elif pos_raw == 0:
                data = list(all_texts_dict_pos.values())
                old_data = all_texts_dict_pos
            else:
                raise Exception
            ldamodel = gensim.models.ldamodel.LdaModel.load("new_output/lda_model_normalized_words_freq__common_"+str(fr)+"_topics_"+str(topic)+"_pos_"+str(pos_raw)+"/lda_model_normalized_words_freq__common_"+str(fr)+"_topics_"+str(topic)+"_pos_"+str(pos_raw)+"_common_.mdl")
            print("/lda_model_normalized_words_freq__common_"+str(fr)+"_topics_"+str(topic)+"_pos_"+str(pos_raw)+"_common_.mdl")
            dictionary = corpora.Dictionary(data)
            dictionary.filter_n_most_frequent(fr)
            corpus = [dictionary.doc2bow(doc) for doc in data]
            print("log =",ldamodel.log_perplexity(corpus))
            all_topics = dict(ldamodel.print_topics(num_topics=topic, num_words=5))
            for a in all_topics:
                print("topic №"+str(a),": ",  " kwds ", all_topics[a])
            print()
            print()
            print()
            vals = list(old_data.values())
            names = list(old_data.keys())
            for i in range(5):
                print(list(reversed(sorted(ldamodel.get_document_topics(dictionary.doc2bow(vals[i])), key=byName_key)))[:3], names[i])
            print()
            print()
            print("*"*20)
    
    
```
